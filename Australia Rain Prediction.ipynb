```python
year = df['Date'].dt.year
year.value_counts().sort_index()
```

```python
train_df = df[year <= 2015]
test_df = df[year > 2015]
print(f"Train is of size : {train_df.shape}")
print(f"Test is of size : {test_df.shape}")
```

```python
train_df.columns
```

```python
target_variable = 'RainTomorrow'
```

```python
continuous
```

```python
continuous.remove(target_variable)
continuous.remove('RainToday')      # Technically it is categorical
```

```python
# Date is not an important categorical feature for us,
# instead we will be using Month to show seasonality
categorical = [col for col in df.columns if (
    df[col].dtype == object)]

categorical.append('RainToday')
```

```python
print(f"Numerical Columns are :\n{continuous}")
print(f"There are {len(continuous)} numerical variables")
print(f"Categorical Columns are :\n{categorical}")
print(f"There are {len(categorical)} categorical variables")
```

We will use date till 2015 as train and rest as test

```python
input_features = continuous + categorical

len(input_features)
```

```python
train_X = train_df[input_features].copy()
train_Y = train_df[target_variable].copy()

test_X = test_df[input_features].copy()
test_Y = test_df[target_variable].copy()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(random_state=42, max_iter=20, tol=0.005).fit(df[continuous])

train_X[continuous] = imputer.transform(train_X[continuous])
test_X[continuous] = imputer.transform(test_X[continuous])
```

```python
train_X[continuous].isna().sum()
```

```python
train_X[categorical].isna().sum()
```

```python
cat = categorical.copy()
cat.remove('Location')
cat.remove('Month')
cat
```

Imputing Categorical features randomly based on the probability of their occuring features

```python
for col in cat:
    values = train_X[col].value_counts().index.tolist()
    probab = train_X[col].value_counts(normalize=True).values
    train_X[col].replace(np.nan, np.random.choice(a=values, p=probab), inplace=True)
    test_X[col].replace(np.nan, np.random.choice(a=values, p=probab), inplace=True)
    
```

```python
train_X[categorical].isna().sum()
```

```python
train_X.isna().sum()
```

```python
# One - Hot Encoding Categorical data using pd.get_dummies

train_X = pd.get_dummies(train_X, columns = categorical)
test_X = pd.get_dummies(test_X, columns = categorical)
```

```python
# Removing unnecessary RainToday variable created due to One-Hot encoding
train_X.drop(labels ='RainToday_0.0', inplace=True, axis=1)
test_X.drop(labels = 'RainToday_0.0', inplace=True, axis=1)
```

Scaling Numerical features using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
train_X[continuous] = scaler.fit_transform(train_X[continuous])
test_X[continuous] = scaler.fit_transform(test_X[continuous])
```

```python
train_X.head()
```

```python
print(f"Size of Train_X set {train_X.shape}")
print(f"Size of Test_X set  {test_X.shape}")
```

```python
from sklearn.metrics import plot_confusion_matrix, plot_roc_curve, classification_report
```

Creating a helper function to get metrics for any given model

```python
def conf_matrix(model, X_test, y_test, cmap='Blues'):
    plot_confusion_matrix(model, X_test, y_test, cmap=cmap)
    plt.grid()
    plt.show()

def roc_curve_custom(model, X_test, y_test):
    plot_roc_curve(model, X_test, y_test)
    plt.plot([0, 1], [0, 1], color='black', linestyle='--')
    plt.show()
    
def evaluate(model, y_pred, X_train=train_X, X_test=test_X, y_train=train_Y, y_test=test_Y):
    # Confusion Matrix
    print('Confusion Matrix')
    print('-'*53)
    conf_matrix(model, X_test, y_test)
    print('\n') 
    
    # Classification Report
    print('Classification Report') 
    print('-'*53)
    print(classification_report(y_test, y_pred))
    print('\n')
    
    # ROC Curve
    print('ROC Curve')
    print('-'*53)
    roc_curve_custom(model, X_test, y_test)
    print('\n')
    
    # Checking model fitness
    print('Checking model fitness') 
    print('-'*53)
    print('Train score:', round(model.score(X_train, y_train), 4))
    print('Test score: ', round(model.score(X_test, y_test), 4))
    print('\n')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

Creating a baseline with Logistic Regression

```python
logreg = LogisticRegression(solver='liblinear', random_state=42)
logreg.fit(train_X, train_Y)
y_pred = logreg.predict(test_X)

evaluate(logreg, y_pred)
```

Almost 81 % on the test set. The model is bit overfitting. Let's see how it can be improved

Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
logreg_params = {
    'C': [1, 1e8, 1e16],
    'fit_intercept': [True, False],
    'max_iter': [50, 100, 150],
    'random_state': [42]
}

logreg_gs = GridSearchCV(logreg, logreg_params, scoring='accuracy', n_jobs=-1, cv=3)
logreg_gs.fit(train_X, train_Y)
```

```python
logreg_gs.best_params_
```

```python
round(logreg_gs.best_score_, 4)
```

```python
y_pred_logreg_gs = logreg_gs.predict(test_X)
evaluate(logreg_gs, y_pred=y_pred_logreg_gs)
```

Nothing improved :|

Decision Tree

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_X, train_Y)
y_pred_tree = dt.predict(test_X)

evaluate(dt, y_pred=y_pred_tree)
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 7, 11],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 3, 5],
    'random_state': [42]
}

dt_gs = GridSearchCV(dt, param_grid=params, scoring='accuracy', n_jobs=-1, cv=3)
dt_gs.fit(train_X, train_Y)
```

```python
dt_gs.best_params_
```

```python
round(dt_gs.best_score_, 4)
```

```python
y_pred_tree_gs = dt_gs.best_estimator_.predict(test_X)
evaluate(dt_gs.best_estimator_, y_pred=y_pred_tree_gs)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
nn = MLPClassifier(solver='sgd',
                    hidden_layer_sizes=(20, 10, 3), random_state=1)

nn.fit(train_X, train_Y)
y_pred_nn = nn.predict(test_X)
evaluate(nn, y_pred=y_pred_nn)
```
